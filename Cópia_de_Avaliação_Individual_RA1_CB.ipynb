# Questão 1: 
Um investimento inicial de $1000$ Reais rende juros compostos a uma taxa de 8% ao ano. Faça um gráfico que mostre como o valor do investimento varia com o tempo e determine qual será o valor do investimento após 10 anos. Utilize a equação $V = P * (1 + r)^t$, onde $P$ é o valor principal do investimento, $r$ é a taxa de juros anual (em decimal) e $t$ é o tempo decorrido em anos.


```python
import matplotlib.pyplot as plt
import numpy as np

P = 1000  # valor principal do investimento
r = 0.08  # taxa de juros anual (em decimal)
t = np.arange(0, 11)  # tempo decorrido em anos (de 0 a 10 anos)

V = P * (1 + r) ** t

plt.plot(t, V)
plt.xlabel('Tempo (anos)')
plt.ylabel('Valor do investimento (R$)')
plt.title('Valor do investimento ao longo do tempo')
plt.show()
```

# Questão 2
Usando, a seu critério, as bibliotecas do Python, crie um gráfico para cada uma das funções a seguir. No label do gráfico deverá estar a função impressa com qualidade matemática (latex) e uma label dizendo se a função é crescente ou decrescente. 

1. $f(x)= e^x - 4x^2$

2. $f(x)= \log_3(x^2 + 1) - \log_3(x) -12$

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return np.exp(x) - 4*x**2

x = np.linspace(-2, 2, 100)

y = f(x)

plt.plot(x, y)
plt.title(r'f(x) = $e^x - 4x^2$, decrescente')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return np.log(x**2 + 1) / np.log(3) - np.log(x) / np.log(3) - 12

x = np.linspace(0.1, 10, 100)

y = f(x)

# Plotando o gráfico
plt.plot(x, y)
plt.title(r'f(x) = $\log_3(x^2 + 1) - \log_3(x) - 12, crescente')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Questão 3
Usando o Sympy, ou qualquer outro recurso disponível na internet, escreva, com qualidade matemática o passo-a-passo algébrico que leva a determinação dos seguintes limites. Ao final, para cada uma das equações você deverá traçar um gráfico da função. 

1. $\lim_{x \to 0} \frac{\sin(2x)}{x}$

2. $\lim_{x \to \infty} \frac{e^x}{x^3}$

```python
# limx→0 sin(2x)/x = limx→0 2cos(2x)/1 = 2cos(0)/1 = 2
# Portanto, o limite de sin(2x)/x quando x tende a 0 é igual a 2.

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(2*x) / x

x = np.linspace(-10, 10, 1000)
y = f(x)

plt.plot(x, y)
plt.title(r'$\lim_{x \to 0} \frac{\sin(2x)}{x}$')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

```

```python
#limx→∞ ex/x^3 = limx→∞ ex/3x^2 = limx→∞ (ex/6x) / (x/2) = 0/∞ = 0
#Portanto, o limite de e^x/x^3 quando x tende a infinito é igual a 0.

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(x) / x**3

x = np.linspace(-10, 10, 1000)
y = f(x)

plt.plot(x, y)
plt.title(r'$\lim_{x \to \infty} \frac{e^x}{x^3}$')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
